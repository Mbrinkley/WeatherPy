```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

#import api key
from api_keys import api_key 
```

```python
# Incorporate citipy to determine city based on latitude and longitude
from citipy import citipy
```

```python
output_data_file = "output_data/cities.csv"
lat_range = (-90, 90)
lng_range=(-180, 180)
```

Cities List:

```python
# list for cities & lat&long
cities = []
lat_lngs = []
#random lat&long combinations 
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs= np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs= zip(lats, lngs)

#identify closest city for each lng&lat
for lat_lng in lat_lngs:
    city=citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    #unique cities - add to cities list
    if city not in cities:
        cities.append(city)
#display city count
len(cities)
```

Perform API Calls:

```python
#call API 
base_url = "http://api.openweathermap.org/data/2.5/weather?"
url = base_url + "appid=" + api_key #+ "&q=" + city
response = requests.get(url)
```

```python
#create a list of objectives for API/df/scatterplots
city_name = []
country = []
date = []
lat = []
lng = []
cloudiness = []
humidity = []
max_temp = []
wind_speed = []
```

```python
#API time/call
record = 1

#loop through all the cities

print(f"Beginning Data Retrieval")
print(f"-------------------------------")

for city in cities: 
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} of set {record}| {city}")
        #print(f"{city}")
        
        # increase count 
        record= record + 1
        
        time.sleep(1.01)
        
    except:
        print("City not found. Skipping...")
    continue
   
            
```

```python
#dictionary for dataframe
weather_dict = {
    "City": city_name,
    "Country":country,
    "Date":date,
    "Lat":lat, 
    "Lng":lng, 
    "Cloudiness":cloudiness,
    "Humidity": humidity,
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
}
```

```python
#dataframe
weatherpy_df = pd.DataFrame(weather_dict)
weatherpy_df.count()
```

```python
#create and save to output folder 
weatherpy_df.to_csv('output_data/weatherpy_df.csv')
```
